```python
import numpy as np
import pandas as pd
```

```python
# örnek data frame oluşturma
data = np.random.randint(-20,20,450).reshape(50,9)
kolonlar = ["k1","k2","k3","k4","k5","k6","k7","k8","k9"]

df = pd.DataFrame(data, columns=kolonlar)
df.head(10)
```

```python
# frame içerisindeki 2 ye bölünebilenler
df[df % 2 ==0].head()
```

```python
# frame içerisindeki 5 ten büyük olanlar
df[df > 5].head()
```

```python
# frame içerisindeki 5 ten küçük olanlar
df[df < 5].head()
```

```python
# frame içerisindeki k2 sütununda 5 ten küçük olanlar
df[df["k2"] < 5]
```

```python
# k1 sütununa göre filtreleme 
df[df["k1"] % 2 ==0].head(10)
```

```python
# k1 sütununa göre filtreleme ve sadece k1 sütununu gösterme
df[df["k1"] % 2 == 0]["k1"] #.loc[:,"k1"]
```

```python
# k1 ve k3 sütununa göre filtreleme 
df[(df["k1"] > 0) & (df["k3"] < 0) & (df["k7"] % 2 == 0)].loc[:,["k1","k3","k7"]]
```

```python
# k1 veya k3 sütununa göre filtreleme 
df[(df["k1"] < 5) | (df["k3"] < 0)].loc[:,["k1","k3"]]
```

```python
# k2 ve k5 sütunlarını query ile filtreleme
df.query("k2 % 2 ==0 & k5 %2 == 0").loc[:,["k2","k5"]]
```

```python
# k2 veya k5 sütunlarını query ile filtreleme
df.query("not(k2 % 2 ==0 | k5 %2 == 0)").loc[:,["k2","k5"]]
```
