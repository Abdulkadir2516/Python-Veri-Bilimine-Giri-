```python
arr = np.random.randint(10,100,25).reshape(5,5)

df = pd.DataFrame(arr, index=["a","c","d","f","g"], columns=["k1","k2","k3","k4","k5"])

df = df.reindex(["a","b","c","d","e","f","g","ğ"])

new_Column = [np.nan,30,np.nan,45,np.nan,54,12,np.nan]
df["k6"] = new_Column

df
```

```python
df1 = df
df1.dropna() # axis = 0 => satırlar
```

```python
df2 = df
df2.dropna(axis = 1) # axis = 0 => satırlar
```

```python
df3 = df
df3.dropna(axis = 0) # axis = 0 => satırlar
```

```python
df4
```

```python
df4 = df
df4.dropna(how = "any") # axis = 0 => satırlar
```

```python
df4 = df
df4.dropna(how = "all") # axis = 0 => satırlar
```

```python
df4 = df
df4.dropna(subset=["k1","k5"], how = "any") # axis = 0 => satırlar
```

```python
df4 = df
df4.dropna(thresh = 4) # en düşük doğru değer
```

```python
df4 = df
df4.fillna(value = "Not")
df4.fillna(value = 1)
df4["k1"].fillna(value = 5)
```

```python
toplam = df.sum().sum()
ort = toplam / (df.size - df.isnull().sum().sum())

df.fillna(value = round(ort))
```
